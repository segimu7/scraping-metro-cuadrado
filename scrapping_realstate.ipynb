```python
import requests
from bs4 import BeautifulSoup

r = requests.get(r"https://www.metrocuadrado.com/casas/arriendo/medellin/#")
c = r.content
soup = BeautifulSoup(c,"html.parser")

all = soup.find_all("div",{"class":"m_rs_list_item"})
```

```python
l = []
for item in all:
    d = {}
    d["Titulo"] = item.find("h2",{"itemprop":"name"}).text
    d["Descripcion"] = item.find("p",{"itemprop":"description"}).text
    d["Area"] = item.find("div",{"class":"m2"}).text.replace("\n","").replace("Área construida:","")
    d["Precio"] = item.find("span",{"itemprop":"price"}).text
    d["Cuartos"] = item.find("div",{"class":"rooms"}).text.replace("\n","").replace("Hab:","").replace("Hab.","")
    
    try:
        d["Baños"] = item.find("div",{"class":"bathrooms"}).text.replace("\n","").replace("Baños:","").replace("Bañ.","")
    except:
        d["Baños"] = "Na"
        l.append(d) 
    try:
        d["Garajes"] = item.find("div",{"class":"bathrooms garages"}).text.replace("\n","").replace("Garajes:","").replace("\t","")
    except:
        d["Garajes"] = 0
        l.append(d) 
        continue 
    l.append(d)       
            
```

```python
import pandas

df = pandas.DataFrame(l)
df
```

```python
df.to_csv("propiedades.csv")
df.to_html("propiedades.html")
```
